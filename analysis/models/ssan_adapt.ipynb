```python
import json
from pathlib import Path
from IPython.display import Image

import numpy as np


path_to_models = Path('../../etc/models/ssan_adapt/docred')
```

# 1. WO dev results

```python
wo_eval_f1 = []
for model_dir in path_to_models.glob('wo_*'):
    with (model_dir / 'test_results.json').open('r') as file:
        res = json.load(file)
        wo_eval_f1.append(res['f1'])
        
print(np.mean(wo_eval_f1))
print(np.std(wo_eval_f1))
```
